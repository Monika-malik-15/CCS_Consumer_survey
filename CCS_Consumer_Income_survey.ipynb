# # Splitting data into Training and Testing:-

```python
x = df.drop('General Economic Condition – one year ahead', axis=1)  
y = df['General Economic Condition – one year ahead'] 
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


print(f"Training data size: {x_train.shape[0]} samples")
print(f"Testing data size: {x_test.shape[0]} samples")
```

# # Standardizing the data for stability and for smooth performance of various machine learning models

```python
# We should map all the feature variables onto the same scale. It is called feature scaling

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

# # Creating model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression


lr_model=LogisticRegression()
lr_model.fit(x_train, y_train)

lr_model_train_score=lr_model.score(x_train, y_train)
print('Training Score is: ', lr_model_train_score)
lr_model_test_score=lr_model.score(x_test, y_test)
print('Testing score is: ', lr_model_test_score)
```

```python
from sklearn.metrics import classification_report


y_pred_lr = lr_model.predict(x_test)
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.model_selection import GridSearchCV


# Grid of hyperparameters to search:
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularization parameter
    'solver': ['lbfgs', 'liblinear', 'saga'],  # Solver algorithms
    'max_iter': [100, 1000, 10000]  # Maximum number of iterations
}

grid_search = GridSearchCV(estimator=lr_model, param_grid=param_grid, cv=10, scoring='accuracy')

grid_search.fit(x_train, y_train)


best_params = grid_search.best_params_
best_lr_model = grid_search.best_estimator_


best_lr_model_score = best_lr_model.score(x_test, y_test)

print("Best Hyperparameters:", best_params)
print("Test Score with Best Model:", best_lr_model_score)
```

```python
from sklearn.preprocessing import label_binarize

# Assuming xgb_model.predict_proba returns probabilities for each class
y_pred_prob_xgb = xgb_model.predict_proba(x_test)

# Binarize the labels
y_test_bin = label_binarize(y_test, classes=xgb_model.classes_)

# Compute ROC curve and ROC area for each class
fpr_xgb = dict()
tpr_xgb = dict()
roc_auc_xgb = dict()
for i in range(len(xgb_model.classes_)):  
    fpr_xgb[i], tpr_xgb[i], _ = roc_curve(y_test_bin[:, i], y_pred_prob_xgb[:, i])
    roc_auc_xgb[i] = auc(fpr_xgb[i], tpr_xgb[i])
```

```python
plt.figure(1)
lw = 2

# Plot ROC curve and AUC for each class
for i in range(len(xgb_model.classes_)):
    plt.plot(fpr_xgb[i], tpr_xgb[i], lw=lw,
             label='Class {} (AUC = {:.2f})'.format(i, roc_auc_xgb[i]))

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('XGBoost ROC-AUC Curve for Multiclass Classification')
plt.legend(loc="lower right")
plt.show()
```
